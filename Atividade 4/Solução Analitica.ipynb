```python
# eda_analitica.py

import numpy as np
import matplotlib.pyplot as plt

# Constantes
T_amb = 25            # Temperatura ambiente (°C)
T0 = 70               # Temperatura inicial do café (°C)
k = 0.005             # Taxa de resfriamento (1/s)
t = np.linspace(0, 2000, 100)  # Tempo de 0 a 2000 s

# Solução analítica
# T(t) = T_amb + (T0 - T_amb) * exp(-k * t)
T = T_amb + (T0 - T_amb) * np.exp(-k * t)

# Plot
plt.plot(t, T, label='Analítica')
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Resfriamento da Caneca de Café - Solução Analítica')
plt.legend()
plt.show()
```
