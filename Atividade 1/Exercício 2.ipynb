```python
#Atividade 2

import numpy as np
import matplotlib.pyplot as plt

# Funo U(x) = x^4 - x^2 e sua derivada
def U(x):
    return x**4 - x**2

def dU_dx(x):
    return 4 * x**3 - 2 * x

# Algoritmo de gradiente descendente com controle de exploso
def gradient_descent(alpha, epsilon=0.01, max_iter=1000, x0=2.0):
    x = x0
    trajectory = [x]
    for _ in range(max_iter):
        grad = dU_dx(x)
        x_new = x - alpha * grad
        # Controle para evitar overflow
        if abs(x_new) > 1e6:
            print(f"Parando, x ficou muito grande: {x_new}")
            break
        trajectory.append(x_new)
        if abs(x_new - x) < epsilon:
            break
        x = x_new
    return trajectory

# Testar diferentes alphas (apenas valores seguros)
alphas = [0.01, 0.05, 0.1]
colors = ['red', 'green', 'blue']

x_vals = np.linspace(-2, 2, 500)
y_vals = U(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, color='black')

# Plotando trajetrias
for alpha, color in zip(alphas, colors):
    traj = np.array(gradient_descent(alpha))
    plt.plot(traj, U(traj), marker='o', linestyle='-', color=color)

plt.title('Trajetria do gradiente descendente em U(x) = x - x')
plt.xlabel('x')
plt.ylabel('U(x)')
plt.grid(True)
plt.show()

```
