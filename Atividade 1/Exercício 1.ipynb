```python

import numpy as np
import matplotlib.pyplot as plt

# Funo U(x) = x^2 - 1
def U(x):
    return x**2 - 1

# Derivada de U(x)
def dU_dx(x):
    return 2 * x

# Algoritmo de gradiente descendente
# alpha: taxa de aprendizado, determina o tamanho do passo que damos na direo do gradiente
#epsilon: tolerncia, determina quando parar o algoritmo
def gradient_descent(alpha=0.1, epsilon=0.01, max_iter=1000, x0=5):
    x = x0
    trajectory = [x]
    for i in range(max_iter):
        grad = dU_dx(x)
        x_new = x - alpha * grad
        trajectory.append(x_new)
        if abs(x_new - x) < epsilon:
            break
        x = x_new
    return trajectory

# Parmetros iniciais
alpha = 0.9
epsilon = 0.02
max_iter = 1000
x0 = 5  # Ponto inicial

# Executar o gradiente descendente
trajectory = gradient_descent(alpha, epsilon, max_iter, x0)

# Preparar dados para o grfico
x_vals = np.linspace(-6, 6, 400)
y_vals = U(x_vals)
trajectory = np.array(trajectory)
trajectory_y = U(trajectory)

# Plotar a funo e a trajetria
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='U(x) = x - 1', color='blue')
plt.plot(trajectory, trajectory_y, marker='o', color='red', label='Trajetria da partcula')
plt.title('Gradiente Descendente na funo U(x) = x - 1')
plt.xlabel('x')
plt.ylabel('U(x)')
plt.legend()
plt.grid(True)
plt.show()
```
